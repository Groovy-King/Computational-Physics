```python
# pyright: reportGeneralTypeIssues = false
import sympy as smp
from scipy.optimize import fsolve

l1 = 1.4710 * 10**11
v1 = 3.0287 * 10**4

M = 1.9891 * 10**(30)
G = 6.67 * 10**(-11)

L = l1 * v1
E = v1**2 / 2 - G * M / l1

# Solving for l2, v2
def orbit_parameters(X):
    x, v = X
    eq1 = x*v - L
    eq2 = v**2 / 2 - G*M/x - E
    return [eq1, eq2]

ecc = smp.sqrt(1 + 2 * E * L**2 / (G**2 * M**2))
factor = smp.Float( (1 + ecc) / (1 - ecc) )
initial_guess = [l1*factor, v1/factor]
print(initial_guess)
solution = fsolve(orbit_parameters, initial_guess)
print(solution)
orbit_parameters(solution)
```

```python
a = (l1 + solution[0]) / 2
T = smp.sqrt(4 * smp.pi**2 * a**3 / (G * M))
smp.N(T / (365 * 24 * 3600))
```
